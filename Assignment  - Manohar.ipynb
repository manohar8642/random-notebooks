```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import re
```

```python
def getAndParseURL(url):
    try:
        result = requests.get(url)
        soup = BeautifulSoup(result.text, 'lxml')
        return(soup)
    except:
        print("URL Not Accessible")
```

```python
main_url = "https://www.imdb.com/"
movies_list = pd.read_csv('movies.csv')

list=[]
for i in range(0,movies_list.shape[0]):
    l=[]
    url = main_url + 'title/' + movies_list['IMDb Title Codes'][i] + '/'
    print('----------------------',i,'----------------------')
    print(url)
    soup = getAndParseURL(url)
    try:
        movie_name = soup.find('div', class_ = 'title_wrapper').h1.text.split('(')[0].strip()
        print(movie_name)
        l.append(movie_name)
    except:
        None
    title_id = movies_list['IMDb Title Codes'][i]
    print(title_id)
    l.append(title_id)
    try:
        budget = [x.text.split(':')[1].split()[0] for x in soup.find('div', class_ = 'article', id = 'titleDetails').find_all('div', class_ = 'txt-block') if x.text.split(':')[0].strip() == 'Budget'][0]
        print(budget)
        l.append(budget)
    except:
        None
    try:
        gross_US = [x.text.split(':')[1].split()[0] for x in soup.find('div', class_ = 'article', id = 'titleDetails').find_all('div', class_ = 'txt-block') if x.text.split(':')[0].strip() == 'Gross USA'][0]
        print(gross_US)
        l.append(gross_US)
    except:
        None
    try:
        runtime = [x.text.split(':')[1].split('\n')[1] for x in soup.find('div', class_ = 'article', id = 'titleDetails').find_all('div', class_ = 'txt-block') if x.text.split(':')[0].strip() == 'Runtime'][0]
        print(runtime)
        l.append(runtime)
    except:
        None
    try:
        url2 = main_url + 'title/' + movies_list['IMDb Title Codes'][i] + '/' + 'releaseinfo?ref_=tt_dt_dt'
        soup2 = getAndParseURL(url2)
        release_dt = [i.find('td', class_ = 'release-date-item__date').text for i in soup2.find_all('tr', class_ = 'ipl-zebra-list__item release-date-item') if i.a.text.split()[0] == 'USA'][-1]
        print(release_dt)
        l.append(release_dt)
    except:
        None
    try:
        url3 = main_url + 'title/' + movies_list['IMDb Title Codes'][i] + '/' + 'parentalguide?ref_=tt_stry_pg'
        soup3 = getAndParseURL(url3)
        MPAA_rating = [i.text.split(':')[1] for i in soup3.find('table').find('tr', class_="ipl-zebra-list__item" ,id="certifications-list").find_all('a') if i.text.split(':')[0]=='United States'][0]
        print(MPAA_rating)
        l.append(MPAA_rating)
    except:
        None
    try:
        genre = [i.get('href').split('=')[1].split('&')[0] for i in soup.find('div', class_ = 'subtext').find_all('a') if i.get('href').startswith('/s')]
        print(genre)
        l.append(genre)
    except:
        None
    try:
        directors = [i.text for i in soup.find_all('div', class_ = 'credit_summary_item')[0].find_all('a')]
        print(directors)
        l.append(directors)
    except:
        None
    try:
        actors = [i.text for i in soup.find_all('div', class_ = 'credit_summary_item')[2].find_all('a')][:-1]
        print(actors)
        l.append(actors)
    except:
        None
    try:
        rating = soup.find('div', class_ = 'ratingValue').text.strip().split('/')[0]
        print(rating)
        l.append(rating)
    except:
        None
    list.append(l)
```

```python
df = pd.DataFrame(list, columns=['Movie.Name','TitleID','Budget','Gross_US','Runtime','US_Release_Date','MPAA_rating','Genre','Directors','Actors','Rating'])
```

```python
df.to_csv('imdb_scrape.csv')
```
